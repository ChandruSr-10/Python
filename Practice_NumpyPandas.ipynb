```python
#numerical analysis on a single array
a=np.array([2,3,5,6,8])

print("adding 1 to each number from rge list",a+1)

print("subtracting 1 to each number from rge list",a-1)

print("multiply 5 to each element from the list",a*5)

print("square of each element from the list",a**2)
```

Generating random numbers

```python
np.arange(1,20,1)
#from 1 to 20 we will get equally spaced numbers (step size of 1)
```

```python
np.arange(1,20,3)
#each number will have an equal difference of 3
```

 linspace() function in Python's NumPy library that allows you to create evenly spaced arrays with a specified number of elements

```python
np.linspace(1,10,50)
#here 50 is the number of elements we need between 1 to 10 (10 is not exclusive)
```

```python
np.linspace(1,10,100)
```

```python
#generating a random float number between 0 and 1(1 willl be exclusive)
num=np.random.rand()
print(num)
```

```python
int_num=np.random.randint(0,10) #random numbers between 1 to 10
print(int_num)
```
