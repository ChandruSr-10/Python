```python
df=data.drop(['Loan_ID'],axis=1)
```

```python
df.head()
```

```python
for col in df.columns:
  df[col]=df[col].fillna(df[col].mean())
#to remove the null values with mean
```

```python
df.isna().sum()
```

```python
#now in the data there are no null values
```

```python
#data loading and data preprocessig is done
```

```python
#dividing the data to check the loan approved or not
x=df.drop(['Loan_Status'], axis=1) #independeant variable
y=df['Loan_Status'] #dependant variable
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.79)
```

```python
#importing logistic regression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(y_test,y_hat)
```
